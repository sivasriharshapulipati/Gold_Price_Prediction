```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

```python
#loading the csv data to pandas 
gold_data = pd.read_csv('/content/gld_price_data.xls')
```

```python
gold_data.head()
```

```python
# print last 5 rows
gold_data.tail()
```

```python
#getting info about data 
gold_data.info()
```

```python
#checking no of missing values
gold_data.isnull().sum()
```

```python
#getting the statistical measures of the data
gold_data.describe()
```

```python
correlation = gold_data.corr()
```

```python
#const heat map
plt.figure(figsize = (8,8))
sns.heatmap(correlation , cbar=True , square=True,fmt = '.1f',annot =True,annot_kws = {'size':8},cmap = 'Blues' )
```

```python
#correlation values of GLD
print(correlation['GLD'])
```

```python
#checking the distridistribution of the GLD price
sns.distplot(gold_data['GLD'],color ='green')
```

```python
#splitting the features and target
X =gold_data.drop(['Date','GLD'],axis = 1)
Y = gold_data['GLD']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state = 2)
```

```python
regressor = RandomForestRegressor(n_estimators=100)
```

```python
regressor.fit(X_train , Y_train)
```

```python
#Model evaluation
#using test dat 
test_data_prediction = regressor.predict(X_test)
```

```python
#R squared error
error_score = metrics.r2_score(Y_test,test_data_prediction)
print("R squared error :",error_score)
```

```python
#compare actual values and predicted values in a plot
Y_test = list(Y_test)
```

```python
plt.plot(Y_test,color='blue',label = 'actual value')
plt.plot(test_data_prediction , color= 'green' , label = 'Predicted Values')
plt.title('actaual prive vs predicted price')
plt.xlabel('no of values')
plt.ylabel('GLD prices')
plt.legend()
plt.show()
```
